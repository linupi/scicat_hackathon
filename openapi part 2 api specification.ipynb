````python
import scicat_new_be_pack
from pprint import pprint
from datetime import datetime
import json
````

````python
# here: let's use the package version of the new backend

configuration = scicat_new_be_pack.Configuration(
    host = "http://35.205.92.39", 
)
````

````python
# authenticate ... dirty workaround solution to fix api problem for now
access_token=None 

with scicat_new_be_pack.ApiClient(configuration) as api_client:
    api_instance = scicat_new_be_pack.AuthApi(api_client)
    credentials_dto = scicat_new_be_pack.CredentialsDto(username = "admin", password = "2jf70TPNZsS") # CredentialsDto | 

    try:
        api_response = api_instance.auth_controller_login(credentials_dto)
        pprint(api_response)
        access_token=api_response["access_token"]    #store token in variable to use it later...
    except ApiException as e:
        print("Exception when calling AuthApi->auth_controller_login: %s\n" % e)
````

````python
dummy_data={"owner_group":"None",
"access_groups":"None",
"creation_location":"somewhere", 
"principal_investigator":"someone",
"dataset_name":"test " + datetime.now().strftime("%Y-%m-%d-%H_%M_%S"),
"type":"raw",
"is_published":False,   
"creation_time":str(datetime.now().isoformat()), 
"source_folder":"None",
"owner":"None",
"contact_email":"a@b.de",
"scientific_metadata":{"test":1.0, "test2":{"value":17,"unit":"mA"}}
}


with scicat_new_be_pack.ApiClient(configuration) as api_client:
    api_client.configuration.access_token=access_token
    api_instance = scicat_new_be_pack.DatasetsApi(api_client)
    create_dataset_dto = scicat_new_be_pack.CreateDatasetDto(**dummy_data) 
    try:
        # another tmp hack is to add _preload_content=False to avoid validation of the returned data in the python module
        api_response = api_instance.datasets_controller_create(create_dataset_dto,_preload_content=False)
        resp = json.loads(api_response.data)
        pprint(resp)
    except Exception as e:
        print("Exception when calling DatasetsApi->datasets_controller_create: %s\n" % e)
    
# creation_location is not part of the data model of openapi so lets try in the next attempt without it
````

basically the current OpenAPI specifications request the following fields for a minium dataset:

```
owner_group, access_groups, created_by, updated_by, owner, owner_email, orcid_of_owner, contact_email, source_folder, source_folder_host, size, packed_size, number_of_files, number_of_files_archived, creation_time, type, validation_status, keywords, description, dataset_name, classification, license, version, is_published, history, datasetlifecycle, created_at, updated_at, instrument_id, techniques, shared_with, local_vars_configuration
```

which is not even in line with what the backend that generated the specification expects.
